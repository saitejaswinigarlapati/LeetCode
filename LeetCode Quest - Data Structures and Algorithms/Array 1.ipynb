#### Q1. Concatenation of Array

```python
from typing import List

class Solution:
    def getConcatenation(self, nums: List[int]) -> List[int]:
        n=len(nums)
        ans=[0]*2*n
        for i in range(n):
            ans[i]=nums[i]
            ans[i+n]=nums[i]
        return ans

s=Solution()
nums=[1,2,1]
print(s.getConcatenation(nums))
```

#### Q2. Shuffle the Array

```python
class Solution:
    def shuffle(self, nums: List[int], n: int) -> List[int]:
        l=nums[:n]
        r=nums[n:]
        ans=[]
        for i in range(n):
            ans.append(l[i])
            ans.append(r[i])
        return ans
s=Solution()
nums=[2,3,5,4,1,7]
n=3
print("Before shuffle: ",nums)
print("After shuffle: ",s.shuffle(nums,n))
```

#### Q3. Max Consecutive Ones

```python
class Solution:
    def findMaxConsecutiveOnes(self, nums: List[int]) -> int:
        c=0
        p=0
        for i in nums:
            if i ==0:
                p=max(p,c)
                c=0
            else:
                c+=1
        return max(c,p)

s=Solution()
nums=[1,1,0,1,1,1]
print("Number of consecutive 1's: ",s.findMaxConsecutiveOnes(nums))
```
