##### Q1. Final Prices With a Special Discount in a Shop

```python
from typing import List

class Solution:
    def finalPrices(self, prices: List[int]) -> List[int]:
        stack = []
        answers = prices[:]   # make a copy
        n = len(prices)

        for i in range(n - 1, -1, -1):
            while stack and prices[stack[-1]] > prices[i]:
                stack.pop()

            if stack:
                answers[i] = prices[i] - prices[stack[-1]]

            stack.append(i)

        return answers


s=Solution()

prices=[8,4,6,2,3]
print(f"prices after discount: ",s.finalPrices(prices))

## Test Cases

# Input: prices = [8,4,6,2,3]
# Output: [4,2,4,2,3]


# Input: prices = [1,2,3,4,5]
# Output: [1,2,3,4,5]

# Input: prices = [10,1,1,6]
# Output: [9,0,1,6]
```

##### Q2 Daily Temperatures

```python
from typing import List

class Solution:
    def dailyTemperatures(self, temperatures: List[int]) -> List[int]:
        st = []
        res = [0] * len(temperatures)

        for i in range(len(temperatures)):
            while st and temperatures[i] > temperatures[st[-1]]:
                idx = st.pop()
                res[idx] = i - idx
            st.append(i)
        
        return res
    
s=Solution()
temperatures = [73,74,75,71,69,72,76,73]
print(s.dailyTemperatures(temperatures))
```
