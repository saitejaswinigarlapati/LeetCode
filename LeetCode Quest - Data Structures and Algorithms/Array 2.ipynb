#### Q1. Set Mismatch

```python
from typing import List

class Solution:
    def findErrorNums(self, nums: List[int]) -> List[int]:
        n=len(nums)
        s=set(nums)
        s_sum=sum(s)
        total=int((n*(n+1))/2)
        duplicate=sum(nums)-s_sum
        missing=total-s_sum
        return [duplicate,missing]
    
n1=[1,2,2,4]
n2=[1,1]
n3=[2,2]

s=Solution()
print("n1: ",s.findErrorNums(n1))
print("n2: ",s.findErrorNums(n2))
print("n2: ",s.findErrorNums(n3))
```

#### Q2. How Many Numbers Are Smaller Than the Current Number

```python
class Solution:
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:
        nums_sorted=sorted(nums)
        rank={}
        for i,num in enumerate (nums_sorted):
            if num not in rank:
                rank[num]=i
        return [rank[num] for num in nums]

s=Solution()

nums1 = [8,1,2,2,3]
nums2 = [6,5,4,8]
nums3 = [7,7,7,7]
print("nums1: ",s.smallerNumbersThanCurrent(nums1))
print("nums2: ",s.smallerNumbersThanCurrent(nums2))
print("nums2: ",s.smallerNumbersThanCurrent(nums3))
```

#### Q3. Find All Numbers Disappeared in an Array

```python
class Solution:
    def findDisappearedNumbers(self, nums: List[int]) -> List[int]:
        nums_set=set(nums)
        missing = [i for i in range(1, len(nums)+1) if i not in nums_set]
        return missing
    

nums1 = [4,3,2,7,8,2,3,1]
nums = [1,1]

s=Solution()

print("nums1: ",s.findDisappearedNumbers(nums1))
print("nums2: ",s.findDisappearedNumbers(nums2))
```
