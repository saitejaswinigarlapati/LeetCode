Q1. Build an Array With Stack Operations

Input: target = [1,3], n = 3

Output: ["Push","Push","Pop","Push"]


Input: target = [1,2,3], n = 3

Output: ["Push","Push","Push"]


Input: target = [1,2], n = 4

Output: ["Push","Push"]

```python
from typing import List

class Solution:
    def buildArray(self, target: List[int], n: int) -> List[str]:
        res=[]
        i=0
        num=1
        while i< len(target):
            t=target[i]
            if t == num:
                res.append('Push')
                i+=1
            else:
                res.append('Push')
                res.append('Pop')
            num+=1
        return res


s=Solution()
target=[1,3]
n=3
print(s.buildArray(target,n))

## Time Complexity: O(len(target)), Space Complexity: O(1)
```

```python
class Solution:
    def buildArray(self, target: List[int], n: int) -> List[str]:
        res=[]
        stack=[]
        for i in range(1,n+1):
            if stack == target:
                break
            else:
                res.append("Push")
                stack.append(i)
                if i not in target:
                    res.append("Pop")
                    stack.pop()
        return res
s=Solution()
target=[2,3,4]
n=4
print(s.buildArray(target,n))

## Time Complexity: O(nÂ²), Space Complexity: O(n)
```

Q2. Evaluate Reverse Polish Notation

```python
class Solution:
    def calculate(self,a,b,op):
        if op=="+":
            return a+b
        elif op=='-':
            return a-b
        elif op=='*':
            return a*b
        elif op=='/':
            return int(a/b)
        
    def evalRPN(self, tokens: List[str]) -> int:
        stack=[]
        op=['+','-','*','/']
        for i in tokens:
            if i not in op:
                stack.append(int(i))
            else:
                b=stack.pop()
                a=stack.pop()
                res=self.calculate(a,b,i)
                stack.append(res)
        ans=stack[0]
        return ans

s=Solution()
tokens=["2","1","+","3","*"]

print(s.evalRPN(tokens))
```
