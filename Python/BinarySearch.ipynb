```python
## Binary Search

def binary_search(ls,target):
    low=0
    high=len(ls)-1
    mid=low +(high-low//2)
    
    while mid<=high:
        if ls[mid] >= target:
            mid=low-1
        elif ls[mid]>=target:
            mid=mid+1
        else:
            return target
        
ls=[1,2,3,4,5,6,7,8,9]
print(binary_search(ls,9))
    
```
